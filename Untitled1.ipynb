# Which is the most and least cost depreciated vehicle in data ?


```python
data['Depreciation'] = data['Present_Price'] - data['Selling_Price']

most_depreciated_vehicle = data.loc[data['Depreciation'].idxmax()]
least_depreciated_vehicle = data.loc[data['Depreciation'].idxmin()]

print(f"Most cost depreciated vehicle: {most_depreciated_vehicle['Car_Name']} with depreciation of {most_depreciated_vehicle['Depreciation']}")
print(f"Least cost depreciated vehicle: {least_depreciated_vehicle['Car_Name']} with depreciation of {least_depreciated_vehicle['Depreciation']}")
```
